**TRATAMENTO DE DADOS CATEGORICOS**


```python
df_clean = df_rent.drop(['New','Property Type','Negotiation Type'], axis = 1)
```

```python
from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder()

district_encoder = oe.fit_transform(df_rent[['District']])
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder()

one_hot = ohe.fit_transform(df_clean[['District']])
```

```python
df_one_hot = pd.get_dummies(df_clean['District'])
df = df_clean.drop('District',axis=1)
df = df.join(df_one_hot)
df
```

### TREINAMENTO ###

```python
from sklearn.model_selection import train_test_split
```

```python
y = df['Price']
x = df.loc[:,df.columns != 'Price']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

**REGRESSÃO LINEAR**

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(x_train,y_train)
```

```python
dadosx = x_train.iloc[:6]
dadosy = y_train.iloc[:6]

print('Previsão: ', lr.predict(dadosx))
print('Real: ', dadosy.values)
```

**ARVORE DE DECISÃO**

```python
from sklearn.tree import DecisionTreeRegressor

tr = DecisionTreeRegressor()
tr.fit(x_train,y_train)
```

```python
print('Previsão: ', tr.predict(dadosx))
print('Real: ', dadosy.values)
```

**RANDOMFOREST REGRESSOR**

```python
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor()
rfr.fit(x_train,y_train)
```

**RMSE**

```python
from sklearn.metrics import mean_squared_error
#Regressão linear
preds = lr.predict(x_train)
lin_mse = mean_squared_error(y_train,preds)
lin_rmse =  np.sqrt(lin_mse)
lin_rmse
```

```python
#Arvore de decisão
tpreds = tr.predict(x_train)
tree_mse = mean_squared_error(y_train,tpreds)
tree_rmse =  np.sqrt(tree_mse)
tree_rmse
```

```python
#Floresta aleatória
rpreds = rfr.predict(x_train)
rfr_mse = mean_squared_error(y_train,rpreds)
rfr_rmse =  np.sqrt(rfr_mse)
rfr_rmse
```

**CROSS VALIDATION**

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(tr,x_train,y_train,scoring='neg_mean_squared_error',cv=10)
tree_rmse_scores = np.sqrt(-scores)

def display_scores(scores):
    print('Scores: ',scores)
    print('Media: ',scores.mean())
    print('Desvio Padrão: ',scores.std())

display_scores(tree_rmse_scores)
```

```python
scores = cross_val_score(lr,x_train,y_train,scoring='neg_mean_squared_error',cv=10)
lin_rmse_scores = np.sqrt(-scores)

def display_scores(scores):
    print('Scores: ',scores)
    print('Media: ',scores.mean())
    print('Desvio Padrão: ',scores.std())

display_scores(lin_rmse_scores)
```

```python
scores = cross_val_score(rfr,x_train,y_train,scoring='neg_mean_squared_error',cv=10)
rfr_rmse_scores = np.sqrt(-scores)

def display_scores(scores):
    print('Scores: ',scores)
    print('Media: ',scores.mean())
    print('Desvio Padrão: ',scores.std())

display_scores(rfr_rmse_scores)
```

### AVALIAR E OTIMIZAR MODELO ###

```python
from sklearn.model_selection import GridSearchCV

param_grid = [{'n_estimators':[3,10,30],'max_features':[2,4,6,8]},
              {'bootstrap':[False],'n_estimators':[3,10],'max_features':[2,3,4]}]

randomf = RandomForestRegressor()

grid_search = GridSearchCV(randomf,param_grid, cv=5,scoring='neg_mean_squared_error')

grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_estimator_
```

```python
final_model = grid_search.best_estimator_
fmp = final_model.predict(x_test)

final_mse = mean_squared_error(y_test,fmp)
print(np.sqrt(final_mse))
```
